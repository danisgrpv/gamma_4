Параметры пучков

```python
linspace1 = np.linspace(0.035, 0.1, 25, dtype=np.float32)
linspace2 = np.geomspace(0.11, 3, 25, dtype=np.float32)

beams_energyes = []
beams_energyes.extend(linspace1)
beams_energyes.extend(linspace2)
```

Уровни пропускания в К-крае

```python
linspace3 = np.linspace(0.001, 0.9, 51)
for i, val in enumerate(linspace3):
    linspace3[i] = round(val, 5)
    
levels = []
levels.extend(linspace3)
```

Зависимость минимальной ошибки восстановления от граничной энерии спектра

```python
def minimal_deviations(dev):
    minimal_deviations = []

    for i,val in enumerate(beams_energyes):
        min_value = min(dev[i])
        minimal_deviations.append(min_value)
    
    return minimal_deviations
```

```python
minimal_deviations_ross80 = minimal_deviations(DEV_ross80)
minimal_deviations_edges80 = minimal_deviations(DEV_edges80)
minimal_deviations_ross50 = minimal_deviations(DEV_ross50)
minimal_deviations_edges50 = minimal_deviations(DEV_edges50)
minimal_deviations_ross30 = minimal_deviations(DEV_ross30)
minimal_deviations_edges30 = minimal_deviations(DEV_edges30)
minimal_deviations_ross40 = DEV_ross40
```

```python
fig, ax = plt.subplots(figsize=(9, 7))

ax.plot(beams_energyes, minimal_deviations_edges80, ls='-', marker='D', lw='0.7', ms=5, color='blue', label='Краевые фильтры')
ax.plot(beams_energyes, minimal_deviations_ross80, ls='-', marker='D', lw='0.7', ms=5, color='magenta', label='Фильтры Росса')

# настройки графика
ax.set_xlabel('Энергия электронов, МэВ', fontsize=10)
ax.set_ylabel('Минимальное среднеквадратичное отклонение по спектру, отн. ед', fontsize=10)
ax.tick_params(axis='both', labelsize=10)
# ax.set_xscale('log')
# ax.set_yscale('log')
ax.set_xlim(3e-2, 0.14)
ax.set_ylim(0, 0.150)

box = ax.get_position()
ax.set_position([box.x0, box.y0 + box.height * 0.1, box.width, box.height * 0.9])

# Put a legend below current axis
ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.1), fancybox=True, shadow=False, ncol=5)
ax.grid(True)

# fig.savefig('Сравнение минимальной ошибки двух методов 30 Материал до 150 кэВ.jpg', bbox_inches='tight')
```
