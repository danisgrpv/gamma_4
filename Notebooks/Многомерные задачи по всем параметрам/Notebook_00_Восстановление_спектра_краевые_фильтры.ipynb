Зависимость ошибки восстановления от номера итерации

```python
beam_energy = 0.075
z_number = 50
```

```python
brem = initialization_brem(z_number, beam_energy)
```

```python
z_list = [10*i for i in range(1,10)]
```

```python
RESULTS_z = np.empty((9), dtype=object)
```

```python
st = time.time()

for i, z_num in enumerate(z_list):
    brem = initialization_brem(z_num, beam_energy)
    signals = np.array([simpson_rule(r*brem) for r in EDGE_METHOD_channels])
    result = Gold(matrix, signals, -1, ENERGY_MESH, new_mesh, brem, iteration_number=10000, bound=bound_100_kev, journal=True, weight_mode='value')
    RESULTS_z[i] = result
    print(z_num, round(time.time() - st, 3))
```
